```python
df.drop(columns=["Customer_ID"], inplace=True)  # Remove non-numeric column
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
encoder = LabelEncoder()

# List of categorical columns
categorical_cols = ["Gender", "Location", "Subscription_Type", "Company_Type"]

# Apply label encoding
for col in categorical_cols:
    df[col] = encoder.fit_transform(df[col])

print("Categorical columns converted successfully!")
```

```python

df.dropna(inplace=True)
print("Dropped rows with missing values!")
```

```python
print(df.isnull().sum())
```

```python
#apply SMOTE
from imblearn.over_sampling import SMOTE
from collections import Counter

# Separate features and target variable
X = df.drop(columns=["Churn_Label"])  # Independent variables
y = df["Churn_Label"]  # Target variable

# Apply SMOTE to balance the dataset
smote = SMOTE(sampling_strategy="auto", random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert back to DataFrame
df_balanced = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=["Churn_Label"])], axis=1)

# Display new class distribution
print("Class distribution after SMOTE:", Counter(y_resampled))

# Save balanced dataset
df_balanced.to_csv("balanced_churn_dataset.csv", index=False)
print("Balanced dataset saved successfully!")
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier

# Load dataset
df2 = pd.read_csv("balanced_churn_dataset.csv")

# Rename target column to 'Churn' if it exists as 'Churn_Label'
df2.rename(columns={"Churn_Label": "Churn"}, inplace=True)

# Ensure 'Churn' column exists
df2['Churn'] = df2['Churn'].astype(int)  # Convert to integer if necessary

# Step 1: Handle missing values
df2.fillna(df2.median(numeric_only=True), inplace=True)  # Fill numerical with median
df2.fillna("Unknown", inplace=True)  # Fill categorical with "Unknown"

# Step 2: Encode categorical features
categorical_cols = df2.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df2[col].nunique() <= 5:  # One-hot encode small categories
        df2 = pd.get_dummies(df2, columns=[col], drop_first=True)
    else:  # Label encode large categories
        le = LabelEncoder()
        df2[col] = le.fit_transform(df2[col])

# Step 3: Create new features
df2['Total_Spend'] = df2.filter(like='Payment').sum(axis=1)  # Sum payment-related columns
if 'Last_Interaction' in df2.columns:
    df2['Days_Since_Last_Interaction'] = (pd.to_datetime('today') - pd.to_datetime(df2['Last_Interaction'])).dt.days

# Step 4: Scale numerical features
scaler = StandardScaler()
numerical_cols = df2.select_dtypes(include=['int64', 'float64']).columns.drop('Churn', errors='ignore')
df2[numerical_cols] = scaler.fit_transform(df2[numerical_cols])

# Step 5: Feature Selection using Random Forest
X = df2.drop(columns=['Churn'])  # Features
y = df2['Churn']  # Target variable

model = RandomForestClassifier()
model.fit(X, y)
feature_importance = pd.Series(model.feature_importances_, index=X.columns).sort_values(ascending=False)
important_features = feature_importance[feature_importance > 0.01].index.tolist()

# Ensure 'Churn' is always in the dataset
df2 = df2[important_features + ['Churn']]

# Save processed dataset
df2.to_csv("processed_data.csv", index=False)
print("Feature Engineering Complete. Processed data saved as processed_data.csv")
```
